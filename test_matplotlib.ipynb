```python
x = np.arange(0, 3*np.pi, 0.1) # chia đoạn từ 0 đến 3 pi thành các đoạn dài 0.1
y = np.sin(x) # tính y
plt.plot(x, y)
plt.xlabel('Ox', color = 'red' )
plt.ylabel('Oy', color = 'red')
plt.title('Hàm SIN trong khoảng 0 đến 3pi', color = 'green')
plt.legend(['SIN(x)'])
```

```python
# prepare new data
rng = np.random.RandomState(1) # tạo ra pseudorandom với giá trị hạt giống (seed) là 0
x = rng.randn(100) # tạo ra mảng ngẫu nhiên gồm 100 số ( do seed bằng 1 nên hầu hết các x mới đều năm [-1,1]
                    # tức là độ lệch chuẩn normal là 1
y = rng.randn(100)

fig,ax = plt.subplots()
ax.scatter(x,y) # trục x biểu thị các giá trị trong mảng x và trục y biểu thị các giá trị trong mảng y
plt.show()
```

```python
rng = np.random.RandomState(0) # tạo ra pseudorandom với giá trị hạt giống (seed) là 0
x = rng.randn(100) # tạo ra mảng ngẫu nhiên gồm 100 số
y = rng.randn(100)

sizes = 100* rng.rand(100) 
print(sizes)
fig,ax = plt.subplots()
ax.scatter(x,y, s= sizes, color = "y") # trục x biểu thị các giá trị trong mảng x và trục y biểu thị các giá trị trong mảng y
plt.show()
```

## California city

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('./data/california_cities.csv') # ./ tức là tại folder tổng
cities
```

```python
cities['area_total_km2'].sort_values(ascending = False)
```

```python
#latd = vĩ độ(đường chạy ngang) oy, longd : kinh độ ox
# Extract Latd, Longd
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']
plt.scatter(lon ,lat,
           c= np.log10(population), cmap = 'viridis', # c = color
           s = area, linewidths = 0, alpha= 0.5 )
plt.style.use('seaborn')

plt.colorbar(label = 'log$_{10}$(population)'); # công thức của log10 , colorbar là cột màu
plt.clim(3,7); # limit cho colorbar
```

```python
plt.axis('equal') # để khối kinh độ vĩ độ bằng khố vĩ độ
plt.xlabel('longlatitude')
plt.ylabel('latitude')
plt.scatter(lon ,lat,
           c= np.log10(population), cmap = 'viridis', # c = color
           s = area, linewidths = 0, alpha= 0.5 )
plt.style.use('seaborn')

plt.colorbar(label = 'log$_{10}$(population)'); # công thức của log10 , colorbar là cột màu
plt.clim(3,7); # limit cho colorbar

area_range = [50, 100, 300, 500]
for area in area_range:
    plt.scatter([], [], s=area, c= 'k' , alpha =0.4,
                label = str(area) +' km$^2$')
plt.legend(labelspacing = 1, title ='City Area') # labelspacing : khoảng cách giữa các label
```
